```python
def _setdiag_dense(A, d):
    A.flat[::len(d)+1] = d
```

```python
def _laplacian_dense(graph, normed=False, axis=0):
    m = np.array(graph)
    np.fill_diagonal(m, 0)
    w = m.sum(axis=axis)
    if normed:
        isolated_node_mask = (w == 0)
        w = np.where(isolated_node_mask, 1, np.sqrt(w))
        print(w)
        m /= w
        print(m)
        m /= w[:, np.newaxis]
        print(m)
        m *= -1
        print(m)
        _setdiag_dense(m, 1 - isolated_node_mask)
        print(m)
    else:
        m *= -1
        print(m)
        _setdiag_dense(m, w)
        print(m, w)
    return m, w
```
